```python
###csv
import os
f=open("/content/drive/MyDrive/Maching_Learning_HW3/ML@NTUT-2021-Autumn-ASR/train-toneless_update.csv",'r')
traindata=[]
for train in f.readlines():
  traindata.append(train.rstrip('\n').split(","))
traindata

file_list=[]
train_path="/content/drive/MyDrive/Maching_Learning_HW3/ML@NTUT-2021-Autumn-ASR/train"
for file in os.listdir(train_path):
  if file.endswith(".wav"):
    file_list.append(file)

final_dic={}
count=0
for data in file_list:
  for train in traindata:
    if train[0]==data[:-4]:
      final_dic[data]=train[1]
      count+=1
print(count) 
```

```python
count=0
txt_path="/content/drive/MyDrive/Maching_Learning_HW3/ML@NTUT-2021-Autumn-ASR/train/txt"
for data in final_dic:
  with open(txt_path+data[:-4]+".txt",'w+')as ans:
    ans.write(str(final_dic[data]))
    count+=1
count
```
